# Regression Analysis 
### y = OutagesCount

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df3 = df2.drop(columns = ['Northern Total Daily Sunspots', 'Southern Total Daily Sunspots', 
                          'Longitude', 'Latitude'])
```

```python
df3.columns
```

```python
y = df3['OutagesCount']

X  = df3[['Total Daily Sunspots', 'Earthquake', 'Magnitude',
       'MEI', 'CO2', 'CH4', 'N2O', 'CFC-11', 'CFC-12', 'TSI', 'Aerosols',
       'Temp', 'Year', 'Month']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 42)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns, columns=['Coefficient'])
coeff_df
```

Interpreting the Coefficients:

- Total Daily Sunspots: For each unit increase in Total Daily Sunspots, the OutagesCount is estimated to decrease by approximately 0.292

- Earthquake: For each unit increase in earthquake occurrences, OutagesCount is estimated to decrease by approximately 0.041

- Magnitude: For each unit increase in Magnitude, OutagesCount is estimated to increase by approximately 0.156

- For each unit increase in MEI, OutagesCount is estimated to increase by approximately 0.146

- CO2, CH4, N2O, CFC-11, CFC-12: For each unit increase in these variables, OutagesCount is estimated to decrease (except for CFC-11, which increases) based on the provided coefficients

- Aerosols: For each unit increase in aerosol concentration, the dependent variable is estimated to decrease substantially by approximately 15.813

- TSI (Total Solar Irradiance): For each unit increase in TSI, OutagesCount is estimated to decrease by approximately 0.123

- Temp: For each unit increase in temperature, OutagesCount is estimated to increase by approximately 0.083.

```python
import statsmodels.api as sm
X_constant = sm.add_constant(X_train)
model = sm.OLS(y_train, X_constant).fit()
print(model.summary())
```

```python
# Aerosols: P-value is really low means possible high correlation
# CFC-12
# CO2
```

```python
predictions = lm.predict(X_test)
```

```python
sns.distplot((y_test-predictions),bins=50);
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

# Regression Analysis 

### y = Earthquake

```python
y2 = df3['Earthquake']

X2 = df3[['Total Daily Sunspots', 'Magnitude',
       'MEI', 'CO2', 'CH4', 'N2O', 'CFC-11', 'CFC-12', 'TSI', 'Aerosols',
       'Temp', 'Year', 'Month']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size = 0.4, random_state = 42)
```

```python
lm2 = LinearRegression()
```

```python
lm2.fit(X_train, y_train)
```

```python
print(lm2.intercept_)
```

```python
coeff2_df = pd.DataFrame(lm2.coef_,X2.columns, columns=['Coefficient'])
coeff2_df
```

```python
X2_constant = sm.add_constant(X_train)
model = sm.OLS(y_train, X2_constant).fit()
print(model.summary())
```

# Interpretation

The magnitude of Aerosols coefficient is significant, which potentially mean that there is strong correlation between atmospheric aerosols causing earthquakes. However, with a P-value of 0.308, it may suggest that there is not correlation at 
all and fails to reject null hypothesis.
