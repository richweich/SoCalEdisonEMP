```python
# IGNORE
def find_outliers_IQR(df2):

   q1=df2.quantile(0.25)

   q3=df2.quantile(0.75)

   IQR=q3-q1

   outliers = df2[((df2<(q1-1.5*IQR)) | (df2>(q3+1.5*IQR)))]

   return outliers
```

```python
%matplotlib inline
from pathlib import Path

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge, LassoCV, BayesianRidge
import statsmodels.formula.api as sm

from dmba import regressionSummary, exhaustive_search
from dmba import backward_elimination, forward_selection, stepwise_selection
from dmba import adjusted_r2_score, AIC_score, BIC_score
```

```python
df2.columns
```

```python
predictors = ['Year', 'Month','Earthquake', 'Magnitude', 'OutagesCount',
       'None', 'D0', 'D1', 'D2', 'D3', 'D4', 'Solar Flare Value']

outcome = 'Total Daily Sunspots'

# Create dummies for Fuel_Type
# X = pd.get_dummies(X, columns=['CAUSE.CATEGORY.DETAIL','CAUSE.CATEGORY', 'CLIMATE.REGION'], drop_first=True)

X = df2[predictors]
y = df2[outcome]
print(X.shape)
X.head(1)
```

```python
X.isna().sum()
```

```python
train_X, valid_X, train_y, valid_y = train_test_split(X, y, test_size=0.4, random_state=1)
```

```python
lm = LinearRegression()
lm.fit(train_X, train_y)
```

```python
print('intercept ', lm.intercept_)
print(pd.DataFrame({'Predictor': X.columns, 'coefficient': lm.coef_}))

# print performance measures
regressionSummary(train_y, lm.predict(train_X))
```

```python
pred_y = lm.predict(train_X)

print('adjusted r2 : ', adjusted_r2_score(train_y, pred_y, lm))
```

```python
# Use predict() to make predictions on a new set
lm_pred = lm.predict(valid_X)

result = pd.DataFrame({'Predicted': lm_pred, 'Actual': valid_y,
                       'Residual': valid_y - lm_pred})
#Print the first 20 predictions
print(result.head(20))
```

```python
def train_model(variables):
    model = LinearRegression()
    model.fit(train_X[variables], train_y)
    return model

def score_model(model, variables):
    pred_y = model.predict(train_X[variables])
    # we negate as score is optimized to be as low as possible
    return -adjusted_r2_score(train_y, pred_y, model)

allVariables = train_X.columns
results = exhaustive_search(allVariables, train_model, score_model)

data = []
for result in results:
    model = result['model']
    variables = result['variables']
    AIC = AIC_score(train_y, model.predict(train_X[variables]), model)

    d = {'n': result['n'], 'r2adj': -result['score'], 'AIC': AIC}
    d.update({var: var in result['variables'] for var in allVariables})
    data.append(d)
pd.set_option('display.width', 100)
print(pd.DataFrame(data, columns=('n', 'r2adj', 'AIC') + tuple(sorted(allVariables))))
pd.reset_option('display.width')
```

```python
def train_model(variables):
    model = LinearRegression()
    model.fit(train_X[variables], train_y)
    return model

def score_model(model, variables):
    return AIC_score(train_y, model.predict(train_X[variables]), model)

best_model, best_variables = backward_elimination(train_X.columns, train_model, score_model, verbose=True)

print(best_variables)
```

```python
def train_model(variables):
    if len(variables) == 0:
        return None
    model = LinearRegression()
    model.fit(train_X[variables], train_y)
    return model

def score_model(model, variables):
    if len(variables) == 0:
        return AIC_score(train_y, [train_y.mean()] * len(train_y), model, df=1)
    return AIC_score(train_y, model.predict(train_X[variables]), model)

best_model, best_variables = forward_selection(train_X.columns, train_model, score_model, verbose=True)

print(best_variables)
```

```python
best_model, best_variables = stepwise_selection(train_X.columns, train_model, score_model, verbose=True)

print(best_variables)
```

# Conclusions

Regression model shows that there is correlation between Solar Flare Value and OutagesCount, lmplot shows that the correlation is negative, which is not always bad. This could mean that as the solar flare values increase, the outage counts tend to decrease, or vice versa. 

It is because that there are other indirect factors such as: 

- Solar flares can cause geomagnetic storms, which can lead to disruptions in power grids and infrastructure. However, the relationship between solar flares and power outages might not be straightforward. Factors such as the strength and direction of the geomagnetic storm, the resilience of the power grid, and other local conditions can influence the impact of solar flares on outage counts.

- There could be other factors at play that influence power outages, which may not be directly related to solar flare activity. For example, weather conditions, equipment failures, maintenance schedules, and human errors can all contribute to power outages independently of solar activity.

Model also shows a moderate r-squared and decently higher P-value value which means this means lower correlation, however, once again, this is due to other indirect factors.
